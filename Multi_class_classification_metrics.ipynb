```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer

df = pd.read_csv('train.csv')
df.head()
```

```python
# Create an imputer to fill missing values with the mean
imputer = SimpleImputer(strategy='mean')
```

```python
from sklearn.model_selection import train_test_split
X = df.iloc[:,1:]
X
Y = df.iloc[:,0]
```

```python
Y = df.iloc[:,0]
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import  DecisionTreeClassifier
```

```python
# Drop rows with any missing values in X_train and y_train
X_train = X_train.dropna()
y_train = y_train[X_train.index] # Make sure y_train is aligned with X_train after dropping rows

# Fit the model
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
y_pred1 = lr.predict(X_test)
y_pred2 = dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,confusion_matrix
print("Logistic Regression:")
print("Accuracy:", accuracy_score(y_test, y_pred1))

print("Precision:", precision_score(y_test, y_pred1, average='weighted'))  # or 'micro', 'macro'
print("Recall:", recall_score(y_test, y_pred1, average='weighted'))  # or 'micro', 'macro'
print("F1 Score:", f1_score(y_test, y_pred1, average='weighted'))  # or 'micro', 'macro'
```

```python
print("Logistic Regression Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred1),columns=list(range(0,10)))
```

```python
print("Decision Tree Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred2),columns=list(range(0,10)))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred1))
```
