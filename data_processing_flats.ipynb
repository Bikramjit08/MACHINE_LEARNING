# **PRICE COLUMN :**

```python
df['price'].head()
```

```python
df['price'].shape
```

```python
df = df.dropna(subset=['price']) # Drop rows where 'price' is NaN
```

```python
df['price'].shape
```

```python
df['price'].value_counts()
```

```python
df[df['price'] == 'Price on Request']
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
df.head()
```

```python
def treat_price(x):
    if isinstance(x, float): # Check if x is already a float
        return x
    else:
        try: # Handle potential IndexError
            x = x.split()
            if len(x) > 1 and x[1] == 'Lac':
                return round(float(x[0])/100,2)
            elif len(x) > 0:  # Check if x has at least one element
                return round(float(x[0]),2)
            else:
                return np.nan # Or handle empty price as needed
        except (ValueError, TypeError): # Handle non-numeric price
            return np.nan # Or handle invalid price as needed

```

```python
df['price'] = df['price'].astype(str).apply(treat_price)
```
